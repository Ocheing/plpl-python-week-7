Load and Explore the Dataset

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load Iris dataset from sklearn
try:
    iris = load_iris()
    # Create DataFrame with feature names
    df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
    # Add species as categorical labels
    df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
    print("‚úÖ Dataset loaded successfully.")
except Exception as e:
    print("‚ùå Error loading dataset:", e)
    df = None

# Proceed only if data loaded
if df is not None:
    # Display first few rows
    print("\nüìä First 5 rows of the dataset:")
    print(df.head())

    # Check data types and missing values
    print("\nüìã Dataset info:")
    print(df.info())

    print("\nüîç Missing values in each column:")
    print(df.isnull().sum())

    # Optional: Summary statistics
    print("\nüìà Summary statistics:")
    print(df.describe())

    # Optional: Plot pairplot to visualize distributions
    sns.set(style="whitegrid")
    sns.pairplot(df, hue="species", corner=True)
    plt.suptitle("Pairwise Feature Plots of Iris Dataset", y=1.02)
    plt.show()

    # Note: No missing values to clean, but examples are provided below:
    # df.dropna(inplace=True)  # Drop rows with missing values
    # df.fillna(method='ffill', inplace=True)  # Fill missing values forward
```

Load and Explore the Dataset

```python
# Basic statistics
print("\nDescriptive Statistics:\n", df.describe())

# Group by species and get mean of each feature
grouped = df.groupby('species').mean()
print("\nMean of Features by Species:\n", grouped)

# Identify patterns
print("\nObservation:")
print("‚Üí Versicolor seems to have intermediate values across features.")
print("‚Üí Virginica generally has the largest petal and sepal measurements.")
```

STEP 3. Data Visualization

1. Line Chart ‚Äì Mean petal length by species (simulating trend over a category)

```python
grouped['petal length (cm)'].plot(kind='line', marker='o')
plt.title('Mean Petal Length per Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

2. Bar Chart ‚Äì Average petal length per species

```python
grouped['petal length (cm)'].plot(kind='bar', color='skyblue')
plt.title('Average Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

3. Histogram ‚Äì Distribution of sepal length

```python
plt.hist(df['sepal length (cm)'], bins=20, color='green', edgecolor='black')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

4. Scatter Plot ‚Äì Sepal length vs. petal length

```python
sns.scatterplot(data=df, x='sepal length (cm)', y='petal length (cm)', hue='species')
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend()
plt.show()
```
